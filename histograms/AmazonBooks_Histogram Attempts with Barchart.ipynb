```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AmazonBooks/AmazonBooks_Bestsellers.csv')
bestsellers = df[df['Price'] <= 50]

print(bestsellers.to_string()) 
```

```python
##Instead of using the histplot or barplot function, I'm using a frequency table in a spreadsheet and a barplot

pricefrequency = pd.read_csv('AmazonBooks/AmazonBooks_price frequency table.csv')
print(pricefrequency.to_string()) 
```

```python
sns.set_theme(palette='husl')
sns.color_palette("husl", 8)
```

```python
## Seaborn histplot
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.histplot(ax=ax, data=bestsellers, x="Price")

ax.set_title("Title")
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')
```

```python
## Seaborn barplot

dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.barplot(ax=ax, data=pricefrequency, x="Price", y="Frequency")
```

```python
## Adding bar labels

dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.barplot(ax=ax, data=pricefrequency, x="Price", y="Frequency")
for i in ax.containers:
    ax.bar_label(i,)
```
