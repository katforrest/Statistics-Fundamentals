```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scores = pd.read_csv('Scores Frequency.csv')

print(scores.to_string()) 
```

```python
sns.set_theme(palette='rocket')
sns.color_palette("rocket", 8)
```

```python
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.barplot(ax=ax, data=scores, x="Scores", y="Frequency", width=1)

ax.set_title("Frequency Histogram")
ax.set_xlabel('Scores')
ax.set_ylabel('Frequency')
for i in ax.containers:
    ax.bar_label(i,)
```

```python
##Calculate the relative frequency using division and round to the 2nd decimal place
Total = scores['Frequency'].sum()
print(Total)
scores["Relative_Frequency"] = round((scores["Frequency"] / Total)*100, 2)
```

```python
##I had to calculate the percentage in the dataframe by diving by total and multiplying by 100.
##Then in the barplot, use ax.bar_label and the argument fmt='' to return a string and include the percentage sign. 

dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
ax = sns.barplot(ax=ax, data=scores, x="Scores", y="Relative_Frequency", width=1)

ax.set_title("Relative Frequency Histogram")
ax.set_xlabel('Scores')
ax.set_ylabel('Relative Frequency')
for i in ax.containers:
    ax.bar_label(i,fmt='%s%%')
```
