```python
##Read a csv file into DataFrame.
##Then select a portion of the data based on an operator, call it 'bestsellers'
##In this example, I wanted to limit my dataset to books that sold for under or equal to $50

df = pd.read_csv('AmazonBooks/bestsellers with categories.csv')
bestsellers = df[df['Price'] <= 50]

print(bestsellers.to_string()) 
```

```python
sns.set_theme(palette='husl')
sns.color_palette("husl", 8)
```

```python
##This is the basic seaborn/matplotlib histogram function, using the data from the column 'Price'
sns.displot(bestsellers, x="Price")
```

```python
##Here I changed the size of the bins by specifying the intervals where the new bin would occur
##You can also specify the bin size many other ways. 

sns.displot(bestsellers, x="Price", bins=[5,10,15,20,25,30,40,50])
```

```python
## To get fancy with the figure aesthetics, I had to use more elements of matplotlib objects
## Controll the size of the chart by creating the matplotlib Figure and Axes objects first
## figure and axes are objects in a matplotlib plot. Use plt.subplots to access them.


dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.histplot(ax=ax, data=bestsellers, x="Price")
```

```python
## object "fig" has two axes objects, xlabel and ylabel

print(fig.axes) 
print(type(ax))
```

```python
## Use plt.subplots and ax objects' methods to change the x and y axis labels and add a title:

dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.histplot(ax=ax, data=bestsellers, x="Price")

ax.set_title("Title")
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')
```

```python
## ax is a single "AxesSubplot" object. 

print(fig.axes)
print(type(ax))
```
