```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AmazonBooks/AmazonBooks_Bestsellers.csv')
bestsellers = df[df['Price'] <= 50]

print(bestsellers.head(10)) 
```

```python
sns.set_theme(palette='husl')
sns.color_palette("husl", 8)
```

```python
sns.boxplot(data=bestsellers, y='Price')
```

```python
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.boxplot(ax=ax, data=bestsellers, x="Price", y="Genre")

ax.set_title("Title")
ax.set_xlabel('Price')
ax.set_ylabel('y-label')
```

```python
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.boxplot(ax=ax, data=bestsellers, y="Price", x="Year")

ax.set_title("Title")
ax.set_xlabel('Year')
ax.set_ylabel('Price')
```

```python
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)
sns.boxplot(ax=ax, data=bestsellers, x="User Rating", y="Genre")

ax.set_title("Title")
ax.set_xlabel('User Rating')
ax.set_ylabel('Genre')
```
