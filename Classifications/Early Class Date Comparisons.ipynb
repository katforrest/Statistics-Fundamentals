```python
df = pd.read_csv('./met_sample/classification_frequency.csv')
classifications = df[df['Frequency'] >= 100]

dims = (15, 35)
fig, ax = plt.subplots(figsize=dims)
sns.barplot(ax=ax, data=classifications, x="Frequency", y="Classification")
for i in ax.containers:
    ax.bar_label(i,)
```
